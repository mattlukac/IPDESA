```python
import numpy as np
from supervise import equation
import matplotlib.pyplot as plt 
```

```python
Gauss = equation.Dataset('gaussian')
Gauss.load()
gaussX = Gauss.Eqn.domain()
```

```python
Pois = equation.Dataset('poisson')
Pois.load()
poisX = Pois.Eqn.domain()
```

```python
def plot_Pois():
    x = Pois.Eqn.domain()
    u = Pois.train[0]
    Theta = Pois.train[1]
    
    numRows = 3
    numCols = 3
    plotIdx = np.random.randint(0, 1199, numRows*numCols).reshape(3,3)
    yMax = max([np.max(u[idx]) for idx in plotIdx]) + 0.05
    yMin = min([np.min(u[idx]) for idx in plotIdx]) - 0.05
    fig, ax = plt.subplots(3,3, figsize=(15,15))
    for i in range(numRows):
        for j in range(numCols):
            ax[i, j].plot(x, u[plotIdx[i,j]])
            ax[i, j].set_title('u0 = %.2f, u1 = %.2f, f = %.2f' % (Theta[plotIdx[i,j],0],
                                                         Theta[plotIdx[i,j],1],
                                                         Theta[plotIdx[i,j],2]))
            ax[i, j].set_xticks([])
            ax[i, j].set_ylim(yMin, yMax)
```

```python
def plot_Gauss():
    x = Gauss.Eqn.domain()
    u = Gauss.train[0]
    Theta = Gauss.train[1]
    
    numRows = 3
    numCols = 3
    plotIdx = np.random.randint(0, 1199, numRows*numCols).reshape(3,3)
    yMax = max([np.max(u[idx]) for idx in plotIdx]) + 0.05
    yMin = min([np.min(u[idx]) for idx in plotIdx]) - 0.05
    fig, ax = plt.subplots(3,3, figsize=(15,15))
    for i in range(numRows):
        for j in range(numCols):
            ax[i, j].plot(x, u[plotIdx[i,j]])
            ax[i, j].set_title(r'$\sigma^2$' + ' = %.2f' % (Theta[plotIdx[i,j],0]))
            ax[i, j].set_xticks([])
            ax[i, j].set_ylim(yMin, yMax)
```

```python
plot_Gauss()
```

```python
plot_Pois()
```
